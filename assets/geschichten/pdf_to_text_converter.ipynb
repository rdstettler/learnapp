This script reads the content of the pdf file in the same folder and prints it to the console. Make sure to have the `PyPDF2` library installed to run this script.

```pythonimport PyPDF2

````python
import PyPDF2

pdf_file = '16_Kurzgeschichten.pdf'  # Replace with your PDF file name

with open(pdf_file, 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    text = ''
    for page in reader.pages:
        text += page.extract_text()

print(text)
````
